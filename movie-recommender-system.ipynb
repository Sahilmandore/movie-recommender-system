```python
# convert the above format into : {'Action','Adventure','Fanatsy','Scifiaction'} format 

# the list must be integer therfore we will use ast function literal_eval
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
       l.append(i['name']) 
    return l 
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'].values
```

```python

def convert3(obj):
    l=[]
    counter =0
    for i in ast.literal_eval(obj):
       if counter != 3:
           l.append(i['name']) 
           counter +=1
       else:
        break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L       
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#it is in format of a string to append it with other list we have to convert it in the list format 
```

```python
#transforming the data : the gap bettween words must be remove so that one word must not get split into two tags
```

```python
movies['cast']=movies['cast'].apply(lambda x :[ i.replace(" ","") for i in x] )
```

```python
movies['crew']=movies['crew'].apply(lambda x :[i.replace(" ","") for i in x ])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x ])
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features = 5000,stop_words ='english')
```

```python
 vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
## when we will sort the movies according to its similarities its will not be reliable on index therefore we will need to just 
## index with it as well for that we will use enmuarte function to get (index,similarity) list
```

```python
def recommend(movie):
    movie_indx = df[df['title']==movie].index[0]
    distances = similarity[movie_indx]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
       
```

```python
recommend('Aliens vs Predator: Requiem')
```
